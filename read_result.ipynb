```python
import tensorflow as tf
from tensorflow.python.summary.summary_iterator import summary_iterator
from glob import glob

```

```python
path_to_events_file = "logs/default/events.out.tfevents.1715094316.DESKTOP-IT1GFJV"
```

```python
tags_set = set()

for e in tf.compat.v1.train.summary_iterator(path_to_events_file):
    for v in e.summary.value:
        tags_set.add(v.tag)

for tag in tags_set:
    print(tag)
```

```python
loss_valid = []
loss_hole  = []
loss_prc   = []
loss_style = []
loss_tv    = []
path_to_events_file = glob('./logs/default/*.DESKTOP-IT1GFJV')
for file in path_to_events_file:
    for e in summary_iterator(file):
        for v in e.summary.value:
            if v.tag == 'loss_valid':
                loss_valid.append(v.simple_value)
            elif v.tag == 'loss_hole':
                loss_hole.append(v.simple_value)
            elif v.tag == 'loss_prc':
                loss_prc.append(v.simple_value)
            elif v.tag == 'loss_style':
                loss_style.append(v.simple_value)
            elif v.tag == 'loss_tv':
                loss_tv.append(v.simple_value)

    total_loss = loss_valid[-1] + 6*loss_hole[-1] + 0.05*loss_prc[-1] + 120*loss_style[-1] + 0.1*loss_tv[-1]
    print(total_loss)
```
